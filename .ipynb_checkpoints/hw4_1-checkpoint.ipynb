```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=8856&md5_url=bc025d22ce73e3b2ea4aec375d1675ad')
```

```python
df
```

```python
TITLES = ['赴日研習人數','赴德研習人數','赴法研習人數','赴奧研習人數','赴俄研習人數']
df2 = df[df['類別']==TITLES[0]]
sum(df2['總計']) 
NUMBERS=[]

for i in range(5):
    df2 = df[df['類別']==TITLES[i]]
    NUMBERS.append(sum(df2['總計']))
```

```python
LABELS = ['Japan', 'Germany', 'France', 'Austria', 'Russia']
plt.pie(NUMBERS, labels=LABELS, autopct='%1.1f%%')
plt.axis('equal')
plt.show
```
