```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
sns.set_context('talk')
```

The loss function for detecting the minimun I chose is $\frac{e^{sin(x)}-x}{x^3+2}$

```python
X = np.linspace(-1,10,10000)
Y = (np.exp(np.sin(X))-X)/(X**3+2)
plt.scatter(X,Y,s=1)
plt.show
def y(x):
    return (np.exp(np.sin(x))-x)/(x**3+2)
```

The initial point is chosen to be 0.6.
Note that changing the initial point may cause the change of the result since this function contains two local minimun.

```python

initial_point = 0.6
h = 0.0001
k = 0.01
tolerance = 10**(-5)
steps = []
max_iter = 10**5
itern = 0
while(itern < max_iter):
    itern += 1
    m = (y(initial_point+h)-y(initial_point-h))/(2*h)
    steps.append(initial_point)
    if(m > tolerance):
        initial_point = initial_point - k * m
    elif(m < -tolerance):
        initial_point = initial_point - k * m
    else:
        break
```

The steps where the iteration walked through is plotted red, since the dots are quite dense, you may not clearly see each steps.

```python
steps = np.array(steps)
X = np.linspace(-1,10,10000)
Y = (np.exp(np.sin(X))-X)/(X**3+2)
plt.plot(X,Y)
Y1 = (np.exp(np.sin(steps))-steps)/(steps**3+2)
plt.plot(steps,Y1,c = 'r' )
```
